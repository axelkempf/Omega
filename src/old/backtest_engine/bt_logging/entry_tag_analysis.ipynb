Dieses Notebook analysiert die Entry-Tags, die du im Rahmen von Phase 3.3 geloggt hast.

- Du siehst welche Tags verwendet wurden
- Wie oft sie auftraten
- Und optional, wie gut sie performten


```
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```
# üîÅ Passe diesen Pfad ggf. an
log_path = "logs/entry_logs/Session_Reversion/entry_log.csv"
df = pd.read_csv(log_path)
df.head()
```

```
df_tags = df[df["entry_allowed"] == True].copy()
df_tags = df_tags[df_tags["tags"].notnull()]
df_tags = df_tags[df_tags["tags"] != "[]"]
df_tags.head()
```

```
# üí° eval() ist unsicher bei Fremddaten ‚Äì nur f√ºr eigene Logs verwenden!
df_tags["tags"] = df_tags["tags"].apply(eval)
all_tags = df_tags["tags"].explode()
tag_counts = Counter(all_tags)
tag_counts
```

```
tag_df = pd.DataFrame.from_dict(tag_counts, orient="index", columns=["count"])
tag_df = tag_df.sort_values("count", ascending=False)
tag_df.plot(kind="bar", legend=False, title="üìä Anzahl Signale pro Tag", figsize=(10, 5))
plt.ylabel("Anzahl")
plt.grid(True)
plt.show()
```
